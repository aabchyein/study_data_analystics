```python
database = client['db_medicals']
database
```

```python
collection = database['RecurrenceOfSurgery']
```

```python
cursor = collection.find({})
data_list = list(cursor)
```

```python
df_ROS = pd.DataFrame(data_list)
df_ROS
```
